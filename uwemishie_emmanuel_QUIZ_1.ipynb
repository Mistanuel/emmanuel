```python
selected_item = 'Wine'
```

```python
grouped_data = data.groupby('Item').sum()
```

```python
if selected_item in grouped_data.index:
    # Retrieve the sum for the selected item for the years 2015 and 2018
    production_2015_sum = grouped_data.loc[selected_item, 'Y2015']
    production_2018_sum = grouped_data.loc[selected_item, 'Y2018']
```

```python
print(f"Total {selected_item} produced in 2015:", production_2015_sum)
print(f"Total {selected_item} produced in 2018:", production_2018_sum)
```

```python
# Calculate correlation between 'Element Code' and each year
correlation_with_element_code = {}
for column in data.columns:
    if column.startswith('Y'):
        correlation = data['Element Code'].corr(data[column])
        correlation_with_element_code[column] = correlation

# Find the year with the minimum correlation
least_correlated_year = min(correlation_with_element_code, key=correlation_with_element_code.get)
min_correlation = correlation_with_element_code[least_correlated_year]

print(f"The year with the least correlation with 'Element Code' is {least_correlated_year} with a correlation coefficient of {min_correlation}.")
```
